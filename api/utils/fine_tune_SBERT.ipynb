# Synthetic Data Generation

````python
import re
import json
import jsonlines
import openai
import time
import pandas as pd
from openai import OpenAI
import tqdm

# # OpenAI API Key (Replace with your actual key)
# openai_api_key = "your_api_key_here"

# Setting up OpenAI client
client = OpenAI(api_key=openai_api_key)
chatgpt_model = "gpt-4o-mini-2024-07-18"

# Load CSV File
file_path = "llama3_vocab_with_meanings.csv"
df = pd.read_csv(file_path, delimiter=":")  # Adjust delimiter if needed

# Use '4o-mini_word' and '4o-mini_word_meaning' Columns (Skip NULL values)
df = df.dropna(subset=["4o-mini_word", "4o-mini_word_meaning"])  # Drop rows where either is NULL
words_to_generate = df[["4o-mini_word", "4o-mini_word_meaning"]].values.tolist()

# Define output file
output_file = "triplet_data.jsonl"

# Check already processed words
processed_words = set()
try:
    with jsonlines.open(output_file, mode="r") as reader:
        for obj in reader:
            processed_words.add(obj["word"])
except FileNotFoundError:
    print("🚀 No existing triplet file found. Creating a new one.")

print(f"✅ {len(processed_words)} words already processed. Skipping them.")
def generate_triplets(word, meaning, num_samples=15):
    triplets = []
    iteration = 0  # Counter to prevent infinite loops

    while len(triplets) < num_samples:
        iteration += 1
        remaining_samples = num_samples - len(triplets)

        prompt_template = f"""
        Generate {remaining_samples} semantically related triplets based on the word and its meaning:
        - Word: **{word}**
        - Meaning: {meaning}

        Each triplet should follow this format:
        1. A **neutral** sentence (anchor) using the word.
        2. A **paraphrased version** of the sentence (positive).
        3. A **semantically different** sentence (negative).

        Example:
        anchor: "The sun is shining brightly."
        positive: "It's a sunny day."
        negative: "It is raining heavily."

        Provide responses in JSON format, containing a list of objects where each object has "anchor", "positive", and "negative" fields. Do not include anything other than valid JSON.
        """
        messages = [{"role": "user", "content": prompt_template}]
        
        response = client.chat.completions.create(
            model=chatgpt_model,
            messages=messages,
            temperature=1
        )

        # Extract generated text
        generated_text = response.choices[0].message.content.strip()

        # Parse JSON output correctly
        try:
            text = re.sub(r"^```json\n|\n```$", "", generated_text, flags=re.MULTILINE)
            triplet_batch = json.loads(text)

            # Ensure triplets are well-formed
            valid_triplets = [
                triplet for triplet in triplet_batch
                if all(key in triplet for key in ["anchor", "positive", "negative"])
            ]
            triplets.extend(valid_triplets)

            print(f"✅ {len(valid_triplets)} triplets generated. Total: {len(triplets)}/{num_samples}")

            # If no valid triplets were generated in this iteration, break to avoid infinite loop.
            if len(valid_triplets) == 0:
                print(f"⚠ No valid triplets generated for '{word}' in this iteration. Breaking out.")
                break

        except Exception as e:
            print(f"⚠ Error parsing JSON response for '{word}': {e}")
        
        # Failsafe: break after too many iterations
        if iteration > 10:
            print(f"⚠ Maximum iterations reached for '{word}'. Breaking out.")
            break
        
        time.sleep(1)  # API rate limit handling

    # Fallback: if less than num_samples were generated, fill up by duplicating the last valid triplet (if available)
    if len(triplets) < num_samples:
        print(f"⚠ Only {len(triplets)} triplets generated for '{word}'.")
        if triplets:
            last_triplet = triplets[-1]
            while len(triplets) < num_samples:
                triplets.append(last_triplet)
            print(f"⚠ Filled up to {num_samples} by duplicating the last valid triplet.")
        else:
            print(f"⚠ No valid triplets generated for '{word}'. Returning empty list.")
    return triplets[:num_samples]  # Ensure exactly num_samples are returned


# Resume Processing from Last Saved Word
with jsonlines.open(output_file, mode="a") as writer:
    for word, meaning in tqdm.tqdm(words_to_generate, desc="Processing Words"):
        if word in processed_words:
            print(f"⏭️ Skipping already processed word: {word}")
            continue  # Skip words that already have triplets

        print(f"🚀 Generating triplets for: {word} ({meaning})")
        triplet_data = generate_triplets(word, meaning, num_samples=15)

        # Write to JSONL file
        for triplet in triplet_data:
            writer.write({
                "anchor": triplet["anchor"],
                "positive": triplet["positive"],
                "negative": triplet["negative"],
                "word": word,
                "meaning": meaning
            })

print(f"✅ Triplet data saved to {output_file} for SBERT fine-tuning!")
````

````python
import re
import json
import jsonlines
import openai
import time
import pandas as pd
import asyncio
from openai import OpenAI
import tqdm
import nest_asyncio

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

# OpenAI API Key (Replace with your actual key)
openai_api_key = "your_api_key_here"

# Setting up OpenAI client
client = OpenAI(api_key=openai_api_key)
chatgpt_model = "gpt-4o-mini-2024-07-18"

# Load CSV File
file_path = "llama3_vocab_with_meanings.csv"
df = pd.read_csv(file_path, delimiter=":")  # Adjust delimiter if needed

# Use '4o-mini_word' and '4o-mini_word_meaning' Columns (Skip NULL values)
df = df.dropna(subset=["4o-mini_word", "4o-mini_word_meaning"])
words_to_generate = df[["4o-mini_word", "4o-mini_word_meaning"]].values.tolist()

# Define new output file for batch generation
output_file = "triplet_data_batch.jsonl"

# Check already processed words
processed_words = set()
try:
    with jsonlines.open(output_file, mode="r") as reader:
        for obj in reader:
            processed_words.add(obj["word"])
except FileNotFoundError:
    print("🚀 No existing triplet file found. Creating a new one.")

print(f"✅ {len(processed_words)} words already processed. Skipping them.")

# Async function to generate triplets for one word (ensuring 100 samples)
async def generate_triplets_async(word, meaning, num_samples=100):
    triplets = []
    iteration = 0  # Counter to prevent infinite loops

    while len(triplets) < num_samples:
        iteration += 1
        remaining_samples = num_samples - len(triplets)

        prompt_template = f"""
        Generate {remaining_samples} semantically related triplets based on the word and its meaning:
        - Word: **{word}**
        - Meaning: {meaning}

        Each triplet should follow this format:
        1. A **neutral** sentence (anchor) using the word.
        2. A **paraphrased version** of the sentence (positive).
        3. A **semantically different** sentence (negative).

        Example:
        anchor: "The sun is shining brightly."
        positive: "It's a sunny day."
        negative: "It is raining heavily."

        Provide responses in JSON format, containing a list of objects where each object has "anchor", "positive", and "negative" fields. Do not include anything other than valid JSON.
        """
        messages = [{"role": "user", "content": prompt_template}]
        
        try:
            # Use the asynchronous API call
            response = await client.chat.completions.acreate(
                model=chatgpt_model,
                messages=messages,
                temperature=1
            )
            generated_text = response.choices[0].message.content.strip()
            # Remove markdown code block markers if present
            text = re.sub(r"^```json\n|\n```$", "", generated_text, flags=re.MULTILINE)
            triplet_batch = json.loads(text)

            # Ensure triplets are well-formed
            valid_triplets = [
                triplet for triplet in triplet_batch
                if all(key in triplet for key in ["anchor", "positive", "negative"])
            ]
            triplets.extend(valid_triplets)
            print(f"✅ {len(valid_triplets)} triplets generated for '{word}'. Total so far: {len(triplets)}/{num_samples}")

            # If no valid triplets were generated in this iteration, break to avoid infinite loop.
            if len(valid_triplets) == 0:
                print(f"⚠ No valid triplets generated for '{word}' in this iteration. Breaking to avoid infinite loop.")
                break

        except Exception as e:
            print(f"⚠ Error generating triplets for '{word}': {e}")
        
        # Failsafe: break after too many iterations
        if iteration > 10:
            print(f"⚠ Maximum iterations reached for '{word}'. Breaking out.")
            break
        
        await asyncio.sleep(1)  # API rate limit handling

    return triplets[:num_samples]  # Return exactly 100 samples

# Async main function to process words in batches
async def main():
    batch_size = 5  # Adjust this based on your API rate limits and performance
    words_list = [ (word, meaning) for word, meaning in words_to_generate if word not in processed_words ]
    
    print(f"🚀 Processing {len(words_list)} words in batches of {batch_size}...")
    
    for i in range(0, len(words_list), batch_size):
        batch = words_list[i:i+batch_size]
        tasks = [ generate_triplets_async(word, meaning, num_samples=100) for word, meaning in batch ]
        batch_results = await asyncio.gather(*tasks)
        
        # Write batch results to output file
        with jsonlines.open(output_file, mode="a") as writer:
            for (word, meaning), triplet_data in zip(batch, batch_results):
                print(f"🚀 Writing {len(triplet_data)} triplets for word: {word}")
                for triplet in triplet_data:
                    writer.write({
                        "anchor": triplet["anchor"],
                        "positive": triplet["positive"],
                        "negative": triplet["negative"],
                        "word": word,
                        "meaning": meaning
                    })
        
        # Small delay between batches to be safe with rate limits
        await asyncio.sleep(2)

    print(f"✅ Triplet data saved to {output_file} for SBERT fine-tuning!")

# Run the async main function using nest_asyncio
if __name__ == "__main__":
    asyncio.run(main())
````

# Finetuning on 5 word data

````python
import torch
import jsonlines
from torch.utils.data import DataLoader
from sentence_transformers import (
    SentenceTransformer,
    InputExample,
    losses,
    models,
    SentenceTransformerTrainingArguments,
    SentenceTransformerTrainer,
)
from transformers import AutoTokenizer
from datasets import Dataset, DatasetDict  # Import both classes
import builtins

# Ensure that Dataset and DatasetDict are available globally.
builtins.Dataset = Dataset
builtins.DatasetDict = DatasetDict

# -------------------------------
# 1. Load Pretrained SBERT Model
# -------------------------------
model_path = "/home/justdial/Downloads/Projects/imgRel/app/models_finetuned/all-mpnet-base-v2"
base_model = SentenceTransformer(model_path)  # Loads the transformer model from the local path

# Load the corresponding tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path)

# *** FIX: Resize the model's token embeddings to match the tokenizer vocabulary ***
base_model[0].auto_model.resize_token_embeddings(len(tokenizer))

# Check and print the embedding dimension using the transformer sub-model
embedding_dim = base_model[0].get_word_embedding_dimension()
print("Word embedding dimension:", embedding_dim)

# Create a pooling layer and build a new SBERT pipeline
pooling_model = models.Pooling(embedding_dim)
sbert_model = SentenceTransformer(modules=[base_model, pooling_model])

# -----------------------------------------
# 2. Load Triplet Training Data from JSONL
# -----------------------------------------
train_examples = []
dataset_file = "triplet_data_allowed_words_15.jsonl"

with jsonlines.open(dataset_file, mode="r") as reader:
    for obj in reader:
        # Create an InputExample for each triplet
        train_examples.append(InputExample(texts=[obj["anchor"], obj["positive"], obj["negative"]]))

print("Number of training examples:", len(train_examples))
# Optionally limit the training examples:
train_examples = train_examples[:496]

# Convert list of InputExample objects to a list of dicts for the Hugging Face dataset
train_data = [
    {"anchor": ex.texts[0], "positive": ex.texts[1], "negative": ex.texts[2]}
    for ex in train_examples
]

# Create a Hugging Face dataset from the list
hf_train_dataset = Dataset.from_list(train_data)
print("Hugging Face dataset columns:", hf_train_dataset.column_names)

# ------------------------------------------------
# 3. Define Training Arguments for Fine-Tuning
# ------------------------------------------------
training_args = SentenceTransformerTrainingArguments(
    output_dir="/home/justdial/Downloads/Projects/imgRel/app/models_finetuned/all-mpnet-base-v2_allowed_words_15",
    num_train_epochs=2,
    per_device_train_batch_size=8,
    learning_rate=2e-5,
    warmup_steps=100,
    fp16=True,  # Set to False if your GPU cannot handle FP16
    save_steps=100,
    logging_steps=100,
    run_name="mpnet-base-allowed-words-triplet"
)

# -----------------------------------------------
# 4. Create the Trainer using TripletLoss
# -----------------------------------------------
trainer = SentenceTransformerTrainer(
    model=sbert_model,
    args=training_args,
    train_dataset=hf_train_dataset,  # Use the Hugging Face dataset
    loss=losses.TripletLoss(sbert_model),
)

# 5. Start Training
trainer.train()

# 6. Save the Trained Model
trainer.save_model(training_args.output_dir)
print("✅ Fine-tuning completed. Model saved to", training_args.output_dir)
````

````python
import torch
import jsonlines
from torch.utils.data import DataLoader
from sentence_transformers import (
    SentenceTransformer,
    InputExample,
    losses,
    models,
    SentenceTransformerTrainingArguments,
    SentenceTransformerTrainer,
)
import logging
logger = logging.getLogger(__name__)
from transformers import AutoTokenizer
from datasets import Dataset, DatasetDict  # Import both classes
import builtins

# Ensure that Dataset and DatasetDict are available globally.
builtins.Dataset = Dataset
builtins.DatasetDict = DatasetDict

import jsonlines
from sentence_transformers import InputExample

def update_tokenizer(tokenizer, dataset_file):
    """
    Updates the tokenizer with new words found in the triplet dataset.

    Args:
        tokenizer: The tokenizer to be updated.
        dataset_file: Path to the JSONL file containing triplet data.

    Returns:
        tokenizer: The updated tokenizer.
        train_examples: A list of InputExample instances for training.
    """
    # -----------------------------------------
    # 1. Load Triplet Training Data from JSONL
    # -----------------------------------------
    train_examples = []
    unique_words = set()

    with jsonlines.open(dataset_file, mode="r") as reader:
        for obj in reader:
            # Create an InputExample for each triplet
            train_examples.append(InputExample(texts=[obj["anchor"], obj["positive"], obj["negative"]]))
            # Collect unique words from the dataset
            if "word" in obj:  # Ensure the key exists
                unique_words.add(obj["word"])

    print("Number of training examples:", len(train_examples))
    
    # -----------------------------------------
    # 2. Add New Tokens to the Tokenizer
    # -----------------------------------------
    num_added_tokens = tokenizer.add_tokens(list(unique_words))
    print(f"Added {num_added_tokens} new tokens to the tokenizer.")

    return tokenizer, train_examples

# -------------------------------
# 1. Load Pretrained SBERT Model
# -------------------------------
model_path = "/home/justdial/Downloads/Projects/imgRel/app/models_pretrained"
dataset_file = "triplet_data.jsonl"
output_model_path = "/home/justdial/Downloads/Projects/imgRel/app/models_finetuned/all-mpnet-base-v2_870"
base_model = SentenceTransformer(model_path)  # Loads the transformer model from the local path

# Load the corresponding tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path)
print(f"{len(tokenizer)}: Total tokens in pretrained model tokenizer")
# tokenizer, train_examples = update_tokenizer(tokenizer, dataset_file)
# Error handling for file operations
try:
    tokenizer, train_examples = update_tokenizer(tokenizer, dataset_file)
except FileNotFoundError:
    logger.error(f"File {dataset_file} not found.")
    raise


# *** FIX: Resize the model's token embeddings to match the tokenizer vocabulary ***
base_model[0].auto_model.resize_token_embeddings(len(tokenizer))

# Check and print the embedding dimension using the transformer sub-model
embedding_dim = base_model[0].get_word_embedding_dimension()
print("Word embedding dimension:", embedding_dim)

# Create a pooling layer and build a new SBERT pipeline
pooling_model = models.Pooling(embedding_dim)
sbert_model = SentenceTransformer(modules=[base_model, pooling_model])

# Convert list of InputExample objects to a list of dicts for the Hugging Face dataset
train_data = [
    {"anchor": ex.texts[0], "positive": ex.texts[1], "negative": ex.texts[2]}
    for ex in train_examples
]

# Create a Hugging Face dataset from the list
hf_train_dataset = Dataset.from_list(train_data)
print("Hugging Face dataset columns:", hf_train_dataset.column_names)
# # Data preparation with validation split
hf_dataset_dict = DatasetDict({
    "train": hf_train_dataset,
    "validation": hf_train_dataset.select(range(100))  # Example split
})
# ------------------------------------------------
# 3. Define Training Arguments for Fine-Tuning
# ------------------------------------------------
training_args = SentenceTransformerTrainingArguments(
    output_dir=output_model_path,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    learning_rate=2e-5,
    warmup_steps=100,
    fp16=True,  # Set to False if your GPU cannot handle FP16
    save_steps=100,
    logging_steps=100,
    run_name="mpnet-base-allowed-words-triplet"
)

# -----------------------------------------------
# 4. Create the Trainer using TripletLoss
# -----------------------------------------------
# trainer = SentenceTransformerTrainer(
#     model=sbert_model,
#     args=training_args,
#     train_dataset=hf_train_dataset,  # Use the Hugging Face dataset
#     loss=losses.TripletLoss(sbert_model),
# )
# Training setup with validation
trainer = SentenceTransformerTrainer(
    model=sbert_model,
    args=training_args,
    train_dataset=hf_dataset_dict["train"],
    eval_dataset=hf_dataset_dict["validation"],
    loss=losses.TripletLoss(sbert_model),
    compute_metrics=lambda p: {"accuracy": 0.0}  # Placeholder for metric computation
)

# 5. Start Training
trainer.train()

# 6. Save the Trained Model
trainer.save_model(training_args.output_dir)
print("✅ Fine-tuning completed. Model saved to", training_args.output_dir)
````
